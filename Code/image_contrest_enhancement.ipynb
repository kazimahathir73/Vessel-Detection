## Saving the Enhanced images using Anisotropic Diffusion

```python
import os
import cv2
from skimage.restoration import denoise_tv_chambolle
import matplotlib.pyplot as plt

# Function to apply anisotropic diffusion to an image
def apply_anisotropic_diffusion(image_path, save_path, weight=0.25):
    image = cv2.imread(image_path)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    denoised_image = denoise_tv_chambolle(gray_image, weight=weight)
    denoised_image = cv2.normalize(denoised_image, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)
    filename = os.path.basename(image_path)
    save_name = os.path.join(save_path, filename)
    cv2.imwrite(save_name, denoised_image)

# Folder paths
input_folder = "F:\\Detecting Small Vessels or Boats with Satellite Imagray using YOLOv9\\Datasets\\SSDD Dataset\\BBox_RBox_PSeg_SSDD\\coco_style\\images\\train"
output_folder = "F:\\Detecting Small Vessels or Boats with Satellite Imagray using YOLOv9\Datasets\\SSDD Dataset\\BBox_RBox_PSeg_SSDD\\coco_style\\images\\train_label_img"
os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.endswith(".jpg") or filename.endswith(".jpeg") or filename.endswith(".png"):
        image_path = os.path.join(input_folder, filename)
        apply_anisotropic_diffusion(image_path, output_folder)
        print(f"Processed: {filename}")

print("All images processed and saved successfully.")
```
